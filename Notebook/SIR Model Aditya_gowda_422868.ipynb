### Importing Libraries

```python
import pandas as pd
import numpy as np
import dash
from dash import html
from dash import dcc
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import datetime
import plotly.graph_objects as go
from scipy import optimize, integrate
from dash.dependencies import Input, Output,State
```

```python
world_data = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
world_data_1 = pd.read_csv(world_data)
world_data_1.head()
```

```python
small_data = pd.read_csv(world_data) #For Opting only few countries
small_data.head()
```

```python
time_idx = small_data.columns[4:]
```

```python
data = pd.DataFrame({'date':time_idx})
data.head()
```

```python
country_list = ['Germany', 'Italy', 'US','Australia']
```

```python
for each in country_list:
    data[each] =np.array(small_data[small_data['Country/Region']==each].iloc[:,4::].sum(axis=0))
```

```python
data.head()
```

```python
time_idx = [datetime.strptime(each, "%m/%d/%y") for each in data.date] #converting all the dates into datetime 
time_str= [each.strftime('%Y-%m-%d') for each in time_idx] #converting datetime function to string
```

```python
N0=3000000 #max susceptible population
beta=0.7   # infection spread dynamics
gamma=0.1  # recovery rate


# condition I0+S0+R0=N0
I0=data.Germany[70]
S0=N0-I0
R0=0
```

```python
def SIR_model(SIR,beta,gamma):
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0         
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return([dS_dt,dI_dt,dR_dt])
```

```python
np.array([S0,I0,R0])
```

```python
SIR=np.array([S0,I0,R0])
propagation_rates=pd.DataFrame(columns={'susceptible':S0,
                                        'infected':I0,
                                        'recovered':R0})



for each_t in np.arange(100):
   
    new_delta_vec=SIR_model(SIR,beta,gamma)
   
    SIR=SIR+new_delta_vec
    
    propagation_rates=propagation_rates.append({'susceptible':SIR[0],
                                                'infected':SIR[1],
                                                'recovered':SIR[2]}, ignore_index= True)
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(1, 1)

ax1.plot(propagation_rates.index,propagation_rates.infected,label='infected',color='k')
ax1.plot(propagation_rates.index,propagation_rates.recovered,label='recovered')
ax1.plot(propagation_rates.index,propagation_rates.susceptible,label='susceptible')

ax1.set_ylim(10, 3500000)
ax1.set_yscale('linear')
ax1.set_title('SIR simulation for germany',size=12)
ax1.set_xlabel('no. of days',size=12)
ax1.legend(loc='best',
           prop={'size': 10});
```

```python
ydata = np.array(data.Germany[75:])
t=np.arange(len(ydata))
len(t)
```

```python
I0=ydata[80]
S0=N0-I0
R0=0
```

```python
def SIR_model_t(SIR,t,beta,gamma):
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return dS_dt,dI_dt,dR_dt
```

```python
def fit_odeint(x, beta, gamma):
    return integrate.odeint(SIR_model_t, (S0, I0, R0), t, args=(beta, gamma))[:,1]
```

```python
popt=[0.5,0.2]
fit_odeint(t, *popt)
```

```python
popt, pcov = optimize.curve_fit(fit_odeint, t, ydata)
perr = np.sqrt(np.diag(pcov))
    
print('standard deviation errors : ',str(perr), ' start infect:',ydata[0])
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
```

```python
fitted=fit_odeint(t, *popt)
```

```python
plt.semilogy(t, ydata, 'o')
plt.semilogy(t, fitted)
plt.title("SIR model for Germany cases(Fitted)")
plt.ylabel("Infected  Population ")
plt.xlabel("No. of Days")
plt.show()
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
print("Basic Reproduction Number R0 " , popt[0]/ popt[1])
print("This ratio is derived as the expected number of new infections. The new infections are sometimes called secondary infections from a single infection in a population where all subjects are susceptible")
```

```python
t_initial=100
t_intro_measures=24
t_hold=21
t_relax=21

beta_max=0.085
beta_min=0.04
gamma=0.02
pd_beta=np.concatenate((np.array(t_initial*[beta_max]),
                       np.linspace(beta_max,beta_min,t_intro_measures),
                       np.array(t_hold*[beta_min]),
                        np.linspace(beta_min,beta_max,t_relax),
                       ))
```

```python
SIR=np.array([S0,I0,R0])
propagation_rates=pd.DataFrame(columns={'susceptible':S0,
                                        'infected':I0,
                                        'recovered':R0})
```

```python
for each_beta in pd_beta:
   
    new_delta_vec=SIR_model(SIR,each_beta,gamma)
   
    SIR=SIR+new_delta_vec
    
    propagation_rates=propagation_rates.append({'susceptible':SIR[0],
                                                'infected':SIR[1],
                                                'recovered':SIR[2]}, ignore_index=True)
```

```python
propagation_rates.tail()
```

```python
fig, ax1 = plt.subplots(1, 1)

ax1.plot(propagation_rates.index,propagation_rates.infected,label='infected',linewidth=3)

t_phases=np.array([t_initial,t_intro_measures,t_hold,t_relax]).cumsum()
ax1.bar(np.arange(len(ydata)),ydata, width=0.8,label=' current infected',color='r')
ax1.axvspan(0,t_phases[0], facecolor='b', alpha=0.2,label='no measures')
ax1.axvspan(t_phases[0],t_phases[1], facecolor='b', alpha=0.3,label='hard measures introduced')
ax1.axvspan(t_phases[1],t_phases[2], facecolor='b', alpha=0.4,label='hold measures')
ax1.axvspan(t_phases[2],t_phases[3], facecolor='b', alpha=0.5,label='relax measures')
ax1.axvspan(t_phases[3],len(propagation_rates.infected), facecolor='b', alpha=0.6,label='repead hard measures')

ax1.set_ylim(10, 3*max(propagation_rates.infected))
ax1.set_yscale('log')
ax1.set_title('SIR simulations',size=20)
ax1.set_xlabel('no. of days',size=20)
ax1.legend(loc='best',
           prop={'size': 8});
```

```python
def SIR_model(SIR,beta,gamma):
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return([dS_dt,dI_dt,dR_dt])

def SIR_model_t(SIR,t,beta,gamma):
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return dS_dt,dI_dt,dR_dt

#Function is defined to optimize curve fit

def fit_odeint(x, beta, gamma):
    return integrate.odeint(SIR_model_t, (S0, I0, R0), t, args=(beta, gamma))[:,1] # we only would like to get dI
```

```python
for each in df_all[1:]:
    ydata = np.array(df_input[each])
    t = np.arange(len(ydata))
    N0 = 6000000
    
    I0 = ydata[1]
    S0 = N0 - I0
    R0 = 0
    
    popt, pcov = optimize.curve_fit(fit_odeint, t, ydata, maxfev= 40000000)
    perr = np.sqrt(np.diag(pcov))
    
    fitted=fit_odeint(t, *popt).reshape(-1,1)
    df_input[each +'_fitted'] = fitted
    

df_input
```

```python
app = dash.Dash(__name__)
# 
fig = go.Figure()

app.layout = html.Div([
    html.H1("SIR Model"),
    html.H4("Please select anyone country"),
    
    dcc.Dropdown(id='country_dropdown_sir',
                options=[ {'label': each,'value':each} for each in df_all[1:]],
                value='Germany', #pre-selected country
                multi= False),
    
    dcc.Graph(id = "SIR_model"
    )
])

@app.callback(
    Output('SIR_model', 'figure'),
    [Input('country_dropdown_sir', 'value')])

def SIR_fig(con_input):
    df= df_input
    
    for i in df[1:]:
        data = []
        trace = go.Scatter(x=df.date,
                        y=df[con_input],
                        mode='lines+markers',
                        name = con_input)
        data.append(trace)
        
        trace_fitted = go.Scatter(x=df.date,
                        y=df[con_input +'_fitted'], 
                        mode='lines+markers',
                        name=con_input+'_fitted')
        data.append(trace_fitted)
        
        
            
    return {'data': data,
            'layout' : dict(
                width=1280,
                height=720,
                title= 'SIR model',
                xaxis={'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=14,color="#7f7f7f"),
                      },
                yaxis={'type':"log",
                       'range':'[1.1,5.5]'
                      }
                
            )
        }

if __name__ == '__main__':
    app.run_server(port=4050)
```
